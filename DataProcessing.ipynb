# Visualization 

```python
visualDF = df.copy()
```

```python
visualDF.set_index('Date', inplace=True)
```

```python
#Conver all to float
visualDF['Price'] = visualDF['Price'].str.replace(',', '').astype(float)
visualDF['Open'] = visualDF['Open'].str.replace(',', '').astype(float)
visualDF['High'] = visualDF['High'].str.replace(',', '').astype(float)
visualDF['Low'] = visualDF['Low'].str.replace(',', '').astype(float)
visualDF['Change %'] = visualDF['Change %'].str.replace('%', '').astype(float)
```

```python
visualDF.head()
```

```python
visualDF = visualDF.drop(['Change %'], axis=1)
```

```python
#Convert to pounds
visualDF['Price'] = visualDF['Price'].div(100).round(2)
visualDF['Open'] = visualDF['Open'].div(100).round(2)
visualDF['High'] = visualDF['High'].div(100).round(2)
visualDF['Low'] = visualDF['Low'].div(100).round(2)
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-white')

fig = plt.figure(figsize=(15,5))

ax = plt.subplot(1,1,1)
ax.plot(visualDF['High'], color='red', lw=2, label="Open")
ax.plot(visualDF['High'], color='blue', lw=2, label="High")
ax.plot(visualDF['Low'], color='green', lw=2, label="Low")
ax.plot(visualDF['Price'], color='black', lw=2, label="Close")
ax.set_xlabel('Year')
ax.set_ylabel('Stock Price in Pound Sterling (Â£)')
ax.legend(loc='best')

fig.tight_layout()
```

```python
fig = ax.get_figure()
fig.savefig("Forecast.png")
```

# Data Standardization 

```python
#make a copy to standardize data
standardDF = df.copy()
```

```python
standardDF.Open[1]
```

```python
#Conver all to float
standardDF['Price'] = standardDF['Price'].str.replace(',', '').astype(float)
standardDF['Open'] = standardDF['Open'].str.replace(',', '').astype(float)
standardDF['High'] = standardDF['High'].str.replace(',', '').astype(float)
standardDF['Low'] = standardDF['Low'].str.replace(',', '').astype(float)
standardDF['Change %'] = standardDF['Change %'].str.replace('%', '').astype(float)
```

```python
standardDF.head()
```

```python
#standardize data
standardDF[['Price','Open','High','Low','Change %']] = StandardScaler().fit_transform(standardDF[['Price','Open','High','Low','Change %']])
standardDF.head()
```

```python
#export to CSV
standardDF.to_csv('standardDF.csv')
```
